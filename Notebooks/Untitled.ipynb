```python
from thermostat import thermostat
import datetime
import seaborn
```

```python
c_dt = datetime.datetime(1999,12,28,0,0,0)
tstat = thermostat(schedule_type='default',current_datetime = c_dt)
c_dt
```

```python
DT = []
tstp_heat = []
tstp_cool = []
for _ in range(0,2000):
    c_dt = c_dt + datetime.timedelta(minutes=1)
    DT.append(c_dt)
    if c_dt.hour == 17 and c_dt.minute == 0:
        print('Occupant has manually overridden the thermostat.')
        cool, heat = tstat.manual_override(tstp_heat=70,tstp_cool=75)
        print(cool, heat)
    else:
        cool, heat = tstat.update_output(current_datetime=c_dt)
        # print('update_output')  
        # print(cool, heat)
    tstp_heat.append(heat)
    tstp_cool.append(cool)


seaborn.lineplot(x=DT,y=tstp_heat)
seaborn.lineplot(x=DT,y=tstp_cool)
```
